```python
Correct = []
for i in range(len(choiceFrameAll)):
    if choiceFrameAll.frame[i] == 1:
        if (choiceFrameAll.Choice_SND1[i] == 'm' and (choiceFrameAll.freql[i] <= choiceFrameAll.freqr[i])) or (choiceFrameAll.Choice_SND1[i] == 'z' and (choiceFrameAll.freql[i] >= choiceFrameAll.freqr[i])):
            Correct.append(1)
        else:
            Correct.append(0)
    if choiceFrameAll.frame[i] == 2:
        if (choiceFrameAll.Choice_SND1[i] == 'z' and (choiceFrameAll.freql[i] <= choiceFrameAll.freqr[i])) or (choiceFrameAll.Choice_SND1[i] == 'm' and (choiceFrameAll.freql[i] >= choiceFrameAll.freqr[i])):
            Correct.append(1)
        else:
            Correct.append(0)

choiceFrameAll["correct"] = Correct
```
